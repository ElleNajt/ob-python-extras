# Running tests

``` bash
./run_expect_test.sh
```

``` bash
./run_expect_tests.sh --file test1.org --test test_name1 test_name2
```

# Updating goldens

``` bash
./run_expect_test.sh --update-goldens
```

``` bash
./run_expect_tests.sh --update-goldens --file babel-formatting.org
```

``` bash
./run_expect_tests.sh --update-goldens --file babel-formatting.org --test timer
```

``` bash
./run_expect_tests.sh --update-goldens --file babel-formatting.org --test testing_sessions_set_variable
```

# Creating new tests for a new module

-   Either create a new file, or add use babel-formatting.org.

-   Make a new header for the module

-   Add a source block about the test
